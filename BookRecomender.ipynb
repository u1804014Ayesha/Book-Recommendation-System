```python
import numpy as np  
url = 'https://www.rokomari.com/book/category/265/rokomari-selected-books?sort=SOLD_COUNT_DESC&priceRange=0to16300&discountRange=0to60'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
book_name = []
writer = []
rating = []
review = []
price = []
stock = []
```

```python
book_data = soup.findAll('div', attrs= {'class': 'book-text-area'})
```

```python
for store in book_data:
    name = store.p.text
    book_name.append(name)
    write = store.find('p', class_ = 'book-author').text
    writer.append(write)
    rating.append(4.5)
    reviewno= store.find('span', class_ = 'text-secondary').text.replace('(','').replace(')','')
    #print(reviewno.text)
    #.text.replace('(','').replace(')','')
    review.append(reviewno)
    pr= store.find('p', class_ = 'book-price').text.replace(" TK. ", " ", 3)
    
    print(pr[5:])
    #print(pr.text)
    price.append(pr[5:])
    
    #print(pr[7])
    
    stk= store.find('p', class_ = 'book-status text-capitalize').text 
    stock.append(stk)
```
